# Q2. Scrape the details team India’s international fixtures from bcci.tv.

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = 'https://www.bcci.tv/'
driver.get(url)
```

```python
driver.get(new_url)
```

```python
date_day_tags = driver.find_elements_by_xpath("//span[@class='fixture__date']")
date_month_tags = driver.find_elements_by_xpath("//span[@class='fixture__month']")
date_tags = driver.find_elements_by_xpath("//span[@class='fixture__datetime tablet-only']//strong")
time_tags = driver.find_elements_by_xpath("//span[@class='fixture__time']")
match_tags = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//strong")
place_tags = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span")
series_tags = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
```

```python
intl_fixtures['Date'] = intl_fixtures['Date_day'] +'-' + intl_fixtures['Date_month']+'-2021'
```

# Q5. Scrape the details of trending repositories on Github.com.

```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://github.com/'
driver.get(url)
```

```python
new_link = driver.find_element_by_xpath("//a[@data-ga-click='(Logged out) Header, go to Trending']")
new_url = new_link.get_attribute('href')
driver.get(new_url)
```

```python
repository_tags = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']//a")
```

```python
urls = []
for i in repository_tags:
    urls.append(i.get_attribute('href'))
urls
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
title = []
desc = []
contributors = []
language = []

for i in urls:
    driver.get(i)
    delay =10
    WebDriverWait(driver,delay).until(EC.element_to_be_clickable((By.XPATH,"//li//a[@data-ga-click='Footer, go to about, text:about']")))
    title.append((driver.find_element_by_xpath("//a[@data-pjax='#js-repo-pjax-container']")).text)
    
    try:
        desc_tag = driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        desc.append(desc_tag.text)
    except NoSuchElementException:
        desc.append('-')
    
    try:
        contributors_tag = driver.find_element_by_partial_link_text("Contributors")
        contributors.append(contributors_tag.text.replace("Contributors ",""))
    except NoSuchElementException:
        contributors.append('-')
    language_tags = driver.find_elements_by_xpath("//li[@class='d-inline']//a")
    lang = []
    for j in language_tags:
        lang.append(j.text.replace('\n','-'))
    language.append(lang)
```

```python
df = pd.DataFrame()
df['Repository title'] = title
df['Repository Description'] = desc
df['Contributors Count'] = contributors
df['Languages Used'] = language
df
```

```python
driver.close()
```
